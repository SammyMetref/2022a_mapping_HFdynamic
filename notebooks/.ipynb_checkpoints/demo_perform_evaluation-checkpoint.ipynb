## A - Experiment with no noise

### 1- SETUP CONFIGURATION 

```python
# OI Grid
lon_min = 232.
lon_max = 237.
lat_min = 33.
lat_max = 38.
time_min = numpy.datetime64('2012-02-01T00')
time_max = numpy.datetime64('2012-02-03T00')     
dx = 0.2                                                 # zonal grid spatial step (in degree)
dy = 0.2                                                 # meridional grid spatial step (in degree)
dt = numpy.timedelta64(1, 'h')                           # temporal grid step

simu_start_date = '2012-01-04T00'                  # Nature run initial date

glon = numpy.arange(lon_min, lon_max + dx, dx)
glat = numpy.arange(lat_min, lat_max + dy, dy)
gtime = numpy.arange(time_min, time_max + dt, dt)
```

### 2- Downloading the reference

```python
from google_drive_downloader import GoogleDriveDownloader as gdd

gdd.download_file_from_google_drive(file_id='15qIUdWyEudC_YFgAPocuBQ9uvxInDnIC',  
                                    dest_path='./GD-NR-DC2022a.zip',
                                    unzip=True)
!rm GD-NR-DC2022a.zip


# Nature run SSH for mapping evaluation
path_ref = '../GD-NR-DC2022a/ssh_filt_201*.nc'
dc_ref = xr.open_mfdataset(path_ref, combine='nested', concat_dim='time', parallel=True) #.sortby('time')
dc_ref = dc_ref.sortby('time')

# Select dc_ref sample
dc_ref_sample = dc_ref.sel(time=slice(time_min, time_max)) 
dc_ref_sample = dc_ref_sample.sel(lon=slice(lon_min+0.001, lon_max-0.001))
dc_ref_sample = dc_ref_sample.sel(lat=slice(lat_min+0.001, lat_max-+0.001))
del dc_ref

# Rename reference variables 
ref_var = {
    'lon' : 'lon',
    'lat' : 'lat',
    'time' : 'time',
    'ssh_corr' : 'ssh'
    
} 
dc_ref_sample = dc_ref_sample.rename(ref_var)
```

### 3 - Reading your maps

```python
# Give the path and file name where your reconstructed maps are stored
oi1_maps_path = '../results/my_oi1_maps.nc'

# Read the reconstructed maps from  ../results repository
ds_my_oi1_maps = xr.open_mfdataset(oi1_maps_path, combine='nested', concat_dim='time', parallel=True) 

# Check that your maps have the correct nomenclature which should be similar to
ds_my_oi1_maps
```

### 5 - Perform mapping evaluation

```python
# Eval
rmse_t_oi1, rmse_xy_oi1, leaderboard_nrmse, leaderboard_nrmse_std = rmse_based_scores(ds_my_oi1_maps, dc_ref_sample)
psd_oi1, leaderboard_psds_score, leaderboard_psdt_score  = psd_based_scores(ds_my_oi1_maps, dc_ref_sample)
# Print leaderboard
data = [['OI SWOT no noise', 
         leaderboard_nrmse, 
         leaderboard_nrmse_std, 
         leaderboard_psds_score, 
         leaderboard_psdt_score,
        'Covariances not optimized',
        'quickstart.ipynb']]
Leaderboard = pd.DataFrame(data, 
                           columns=['Method', 
                                    r"$\overline{RMSE_{S}}$", 
                                    r"$\sigma(RMSE_{S})$", 
                                    r'$\lambda_{x}$ (degree)', 
                                    r'$\lambda_{t}$ (days)', 
                                   'Notes',
                                   'Reference'])
print("Summary of the leaderboard metrics:")
Leaderboard
print(Leaderboard.to_markdown())
```

### PLOT EVALUATION SCORES

### Reference and reconstructed maps (at 1 day)

```python
map_c = xr.concat((dc_ref_sample.ssh[24],ds_my_oi1_maps.ssh[24]), dim='experiment')
map_c['experiment'] = ["Ref","No noise"]
map_c.hvplot.contourf(x='lon', y='lat', levels=list(numpy.arange(-0.6,0.6, 0.05)), height=300, width=400, cmap='RdYlGn', subplots=True, by='experiment', clabel='Reconstructed maps[m]')
```

### RMSE-based score timeseries

```python
rmse_c = rmse_t_oi1.expand_dims(dim='experiment')
rmse_c['experiment'] = ["No noise"]
rmse_c.hvplot.line(x='time', y='rmse_t', by='experiment', ylim=(0, 1), cmap=['royalblue'], title='RMSE-based scores')
```

### Spatial RMSE map

```python
rmse_xy_c = rmse_xy_oi1.expand_dims(dim='experiment')
rmse_xy_c['experiment'] = ["No noise"]
rmse_xy_c.hvplot.contourf(x='lon', y='lat', levels=list(numpy.arange(0.,0.75, 0.05)), height=300, width=400, cmap='Reds', subplots=True, by='experiment', clabel='RMSE[m]', title='RMSE map')
```

### PSD-based score

```python
psd_c = psd_oi1.expand_dims(dim='experiment')
psd_c['experiment'] = ["No noise"]
plot_psd_score_v0(psd_c)
```

The PSD-based score evaluates the spatio-temporal scales resolved in the mapping. The resolutions can be defined as the contour where the PSD-score = 0.5, black contour in the figure (i.e., the spatio-temporal scales where the reconstruction SSH error level is 2 times lower than the true SSH signal).
